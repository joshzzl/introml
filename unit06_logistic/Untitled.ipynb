```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

data = np.array([[30,0,0],[50,1,1],[70,1,0],\
               [80,2,1],[100,1,1]])
df = pd.DataFrame(data, columns=['x1','x2','y'])

df_0 = df[df['y']==0]
df_1 = df[df['y']==1]

plt.scatter(df_0['x1'], df_0['x2'], marker='o')
plt.scatter(df_1['x1'], df_1['x2'], marker='^')
plt.xlabel('x1 - Income')
plt.ylabel('x2 - Num website')
plt.grid()
plt.show()
```
